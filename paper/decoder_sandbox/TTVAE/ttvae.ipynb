```python
def ttvae(dat, latent_rate=0.2, runs=10, drop_index=True):
    if drop_index == True:
        trn = pd.read_csv(dat + '_trn.csv').iloc[:, 1:]
        tst = pd.read_csv(dat + '_tst.csv').iloc[:, 1:]
    else:
        trn = pd.read_csv(dat + '_trn.csv')
        tst = pd.read_csv(dat + '_tst.csv')
    trn = trn[trn['y']==1].drop('y', axis=1)
    discrete_columns = trn.columns[trn.dtypes == 'object']

    latent_layer = round(latent_rate * trn.shape[1])
    for i in range(runs):
        ttvae = TTVAE(latent_dim = latent_layer)
        ttvae.fit(trn, discrete_columns, '../TTVAE/ckpt')

        transformed = ttvae.transformer.transform(tst)
        transformed = torch.from_numpy(transformed.astype('float32')).to('cuda:0')

        mu, std, logvar, enc_embed = ttvae.encoder(transformed)
        synthetic_embeddings=mu
        noise = torch.Tensor(synthetic_embeddings).to(ttvae._device)
        ttvae.decoder.eval()
        with torch.no_grad():
            fake, sigmas = ttvae.decoder(noise,enc_embed)
            fake = torch.tanh(fake).cpu().detach().numpy()
        tst_rec = ttvae.transformer.inverse_transform(fake)

        tst_rec.to_csv("../ttvae_data/" + dat + '/' + str(latent_rate) + '_run' + str(i+1) +  ".csv", index=False)
    print("Finished " + dat)
    return None
```
